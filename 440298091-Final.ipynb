```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

```python
def plot(mu,sigma,n,a,b,style,bars):
    e=np.random.normal(mu,sigma,n)
    if style == 0 :
        return b*e+a
    if style == 1 :
        plt.hist(b*e+a,bars)
    if style == 2:
        plt.plot(b*e+a)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plot(0,1,1000,0,1,2,1)
plt.subplot(2,2,2)
plot(0,1,1000,2,1,2,1)
plt.subplot(2,2,3)
plot(0,1,1000,0,2,2,1)
plt.subplot(2,2,4)
plot(0,1,1000,2,2,2,1)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(3,2,1)
plot(0,1,1000,0,1,1,10)
plt.subplot(3,2,2)
plot(0,1,1000,0,1,1,10)
plt.subplot(3,2,3)
plot(0,5,1000,0,1,1,10)
plt.subplot(3,2,4)
plot(0,5,1000,0,1,1,10)
plt.subplot(3,2,5)
plot(2,2,1000,0,1,1,10)
plt.subplot(3,2,6)
plot(2,2,1000,0,1,1,10)
plt.show()
```

```python
def new_plot(mu,sigma,n,style,bars):
    e_new=plot(0,1,1000,0,1,0,10).cumsum()
    if style == 0 :
        return e_new
    if style == 1 :
        plt.hist(e_new,bars)
    if style == 2:
        plt.plot(e_new)        
```

```python
plt.figure(figsize=(10,5))
plt.subplot(3,1,1)
new_plot(0,1,1000,2,10)
plt.subplot(3,1,2)
new_plot(0,1,1000,2,10)
plt.subplot(3,1,3)
new_plot(0,1,1000,2,10)
```

```python
new_plot(0,1,1000,1,100)
```

be nazar miresad nazm khassi vujud nadarad.

```python
def repeat(mu,sigma,n,nn,m):
    lst=[]
    for i in range(nn):
        lst.append(new_plot(0,1,1000,0,100)[m])
    plt.hist(lst)
repeat(0,1,1000,1000,30)    
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
repeat(0,1,1000,1000,10)
plt.subplot(2,2,2)
repeat(0,1,1000,1000,100)
plt.subplot(2,2,3)
repeat(0,1,1000,1000,500)
plt.subplot(2,2,4)
repeat(0,1,1000,1000,900)
plt.show()
```

```python
k,l,a,b=sy.symbols('k,l,a,b')
y1=(k**a)*(l**b);display(y1)
y2=(k**a)*(l**b);display(y2)
firm1 = {a: 1/3, b: 2/3, l: 0.23}
firm2 = {a: 2/3, b: 1/3, l: 0.23}
y1=y1.subs(firm1);display(y1)
y2=y2.subs(firm2);display(y2)
dy1=y1.diff(k);display(dy1)
dy2=y2.diff(k);display(dy2)
f_y1=sy.lambdify(k,y1)
f_y2=sy.lambdify(k,y2)
f_dy1=sy.lambdify(k,dy1)
f_dy2=sy.lambdify(k,dy2)
k=np.linspace(10,1000)
f_y11=f_y1(k)
f_y21=f_y2(k)
f_dy11=f_dy1(k)
f_dy21=f_dy2(k)
plt.plot(k,f_y11,color='blue')
plt.plot(k,f_y21,color='orange')
plt.plot(k,f_dy11,color='green',ls='--')
plt.plot(k,f_dy21,color='red',ls='-.')
plt.show()
```
