```python
file=open('فایل قیمت.csv','r')
prices=file.read()
prices=prices.splitlines()
prices=list(map(int,prices))
```

<p style='color:dark orange;font-size:2.1em;text-align:center;direction:rtl;font-family:B koodak'>ب/میانگین و انحراف معیار و بررسی نوع داده ها</p>

```python
n1=0
n2=0
sum1=0
A=0
for i in range(0,len(prices)):
    sum1+=prices[i]
    a=(prices[i]**2)/250
    A+=a
ave=sum1/len(prices)
var=(A)-(ave**2)
dev=(var)**0.5
print(ave,dev)
lst1=[]
lst2=[]
lst3=[]
lst4=[]
for i in range (0,len(prices)):
    if prices[i]<(ave-2*dev):
        lst1.append(prices[i])
    if (ave+2*dev)>prices[i]>(ave-2*dev):
        lst2.append(prices[i])
    if prices[i]>(ave+2*dev):
        lst3.append(prices[i])
print('Karan Payin: ',lst1)
print('Normal: ',lst2)
print('Karan Bala: ',lst3)
```

<p style='color:dark orange;font-size:2.1em;text-align:center;direction:rtl;font-family:B koodak'>ج/تعریف توابع مورد نظر</p>

```python
def AverageG(x):
    D1=[]
    for i in range(0,len(x)-1):
        d1=x[i+1]-x[i]
        D1.append(d1)
    aveG=[]
    sum1=0
    for i in range(0,14):
        if D1[i]>0:    
            sum1+=D1[i]
    aveg1=sum1/14
    aveG.append(aveg1)
    for j in range(14,len(x)-1):
        if D1[j]>0:
            for i in aveG:
                aveg2=(i*13+D1[j])/14
            aveG.append(aveg2)
        if D1[j]==0 or D1[j]<0:
            for i in aveG: 
                aveg2=(i*13)/14
            aveG.append(aveg2)
    return(aveG)        
def AverageL(y):
    D2=[]
    for i in range(0,len(y)-1):
        d2=y[i+1]-y[i]
        D2.append(d2)
    aveL=[]
    sum1=0
    for i in range(0,13):
        if D2[i]<0:
            sum1+=(-D2[i])
    avel1=sum1/14
    aveL.append(avel1)    
    for j in range(13,len(y)-1):
        if D2[j]<0:
            for i in aveL:
                avel2=(i*13+(-D2[j]))/14
            aveL.append(avel2)
        if D2[j]>0 or D2[j]==0:
            for i in aveL:
                avel2=(i*13)/14
            aveL.append(avel2) 
    return(aveL)
def RSI(z):
    RS=[]
    for i in range(len(AverageG(z))) :
        rs=AverageG(z)[i]/AverageL(z)[i]
        RS.append(rs)
    RSi=[]
    for i in RS:
        rsi=100-(100/(i+1))
        RSi.append(rsi)
    return(RSi)
```

```python
lst_RSI=RSI(prices)
print(lst_RSI)
```
